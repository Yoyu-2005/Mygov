```python
import faiss
import numpy as np
from sentence_transformers import SentenceTransformer
from transformers import pipeline
import gradio as gr
from pyngrok import ngrok
```

```python
embedder = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
log_data = [
    "Error: unable to connect to database",
    "User admin logged in",
    "Disk space low on /dev/sda1",
    "User guest failed login attempt",
    "System rebooted successfully"
]
```

```python
log_embeddings = embedder.encode(log_data)
dimension = log_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(log_embeddings))
```

```python
llm = pipeline("text2text-generation", model="google/flan-t5-base")
```

```python
def chat_with_logs(query):
    query_embedding = embedder.encode([query])
    D, I = index.search(np.array(query_embedding), k=3)
    retrieved = [log_data[i] for i in I[0]]

    context = "\n".join(retrieved)
    prompt = f"Logs:\n{context}\n\nQuestion: {query}\nAnswer:"
    response = llm(prompt, max_length=100)[0]['generated_text']
    return response

chat_ui = gr.Interface(fn=chat_with_logs,
                       inputs=gr.Textbox(lines=2, placeholder="Ask something about the logs..."),
                       outputs="text",
                       title="LOGWISE")
```
